```python
import pandas as pd
export_df  = pd.read_csv("Export_2022_24.csv")
export_df
```

```python
unique_units = export_df["UNIT"].unique()
print(unique_units)
```

```python
unit_frequencies = export_df["UNIT"].value_counts()
sorted_units = unit_frequencies.index.tolist()  # Sorted by frequency
print(unit_frequencies)  # Print the counts for each unique value
```

```python
export_df_cleaned = export_df[(export_df["UNIT"]=="MTS")|(export_df["UNIT"]=="KGS")]
export_df_cleaned
```

```python
import pandas as pd

# Define the volume conversion dictionary
volume_conversion = {
    "KGS": 1,
    "MTS": 1000
}

# Filter rows where UNIT is in the dictionary keys
filtered_df = export_df_cleaned[export_df_cleaned["UNIT"].isin(volume_conversion.keys())]

# Add a new column 'conversion' with the corresponding values from the dictionary
filtered_df["conversion"] = filtered_df["UNIT"].map(volume_conversion)
```

```python
filtered_df
```

```python
# Remove non-numeric characters like '$' from the relevant columns
filtered_df["QUANTITY"] = filtered_df["QUANTITY"].astype(str).str.replace(r"[^\d.]", "", regex=True)
filtered_df["VALUE(INR)"] = filtered_df["VALUE(INR)"].astype(str).str.replace(r"[^\d.]", "", regex=True)
filtered_df["UNIT PRICE"] = filtered_df["UNIT PRICE"].astype(str).str.replace(r"[^\d.]", "", regex=True)

# Convert to numeric after cleaning
filtered_df["QUANTITY"] = pd.to_numeric(filtered_df["QUANTITY"], errors="coerce")
filtered_df["VALUE(INR)"] = pd.to_numeric(filtered_df["VALUE(INR)"], errors="coerce")
filtered_df["UNIT PRICE"] = pd.to_numeric(filtered_df["UNIT PRICE"], errors="coerce")

filtered_df = filtered_df[
    (filtered_df["QUANTITY"] != 0) |
    (filtered_df["VALUE(INR)"] != 0) |
    (filtered_df["UNIT PRICE"] != 0)
    ]

filtered_df
```

```python
# Drop rows where 'QUANTITY', 'VALUE(USD)', or 'UNIT PRICE' is NaN
filtered_df = filtered_df.dropna(subset=["UNIT PRICE"])

# Ensure 'UNIT PRICE' and 'conversion' are numeric for further calculations
filtered_df["UNIT PRICE"] = pd.to_numeric(filtered_df["UNIT PRICE"], errors="coerce")
filtered_df["conversion"] = pd.to_numeric(filtered_df["conversion"], errors="coerce")

# Calculate the normalized prices
filtered_df["normalised prices"] = filtered_df["UNIT PRICE"] / filtered_df["conversion"]

print(filtered_df)

filtered_df = filtered_df.dropna(subset=["QUANTITY", "VALUE(INR)", "UNIT PRICE"])
# Calculate the normalized prices
filtered_df["normalised prices"] = filtered_df["UNIT PRICE"] / filtered_df["conversion"]

print(filtered_df)
```

```python
filtered_df
```

```python
# Calculate Q1, Q3, and IQR
Q1 = filtered_df["normalised prices"].quantile(0.25)
Q3 = filtered_df["normalised prices"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out rows where 'normalised prices' are outliers
filtered_df_no_outliers = filtered_df[
    (filtered_df["normalised prices"] >= lower_bound) &
    (filtered_df["normalised prices"] <= upper_bound)
    ]

filtered_df_no_outliers
```

```python
# Filter out rows where QUANTITY or normalised prices is 0
filtered_df_no_outliers = filtered_df_no_outliers[
    (filtered_df_no_outliers["QUANTITY"] != 0) &
    (filtered_df_no_outliers["normalised prices"] != 0)
    ]

filtered_df_no_outliers
```

```python
# Multiply 'conversion' with 'QUANTITY' and create a new column
filtered_df_no_outliers["converted_quantity"] = (
        filtered_df_no_outliers["conversion"] * filtered_df_no_outliers["QUANTITY"]
)

filtered_df_no_outliers
```

```python
import matplotlib.pyplot as plt

# Calculate Q1, Q3, and IQR for 'converted_quantity'
Q1 = filtered_df_no_outliers["converted_quantity"].quantile(0.25)
Q3 = filtered_df_no_outliers["converted_quantity"].quantile(0.75)
IQR = Q3 - Q1

# Define lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out rows where 'converted_quantity' is an outlier
filtered_df_no_outliers = filtered_df_no_outliers[
    (filtered_df_no_outliers["converted_quantity"] >= lower_bound) &
    (filtered_df_no_outliers["converted_quantity"] <= upper_bound)
    ]

# Plot 'converted_quantity' vs 'normalised prices'
plt.figure(figsize=(10, 6))
plt.scatter(
    filtered_df_no_outliers["normalised prices"],
    filtered_df_no_outliers["converted_quantity"],
    alpha=0.7,
    edgecolor="k"
)
plt.title("Normalised Prices (Outliers Removed) vs Converted Quantity", fontsize=14)
plt.xlabel("Normalised Prices", fontsize=12)
plt.ylabel("Converted Quantity", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Calculate Pearson's correlation
correlation = filtered_df_no_outliers["converted_quantity"].corr(
    filtered_df_no_outliers["normalised prices"]
)

print(f"Pearson's correlation coefficient: {correlation:.4f}")
```

```python
# Convert 'DATE' column to date format
filtered_df_no_outliers["DATE"] = pd.to_datetime(filtered_df_no_outliers["DATE"], errors="coerce")

# Group by 'DATE' and compute required metrics
result_df = filtered_df_no_outliers.groupby("DATE").agg({
    "normalised prices": "mean",   # Mean of normalised prices
    "converted_quantity": "sum"   # Sum of converted_quantity
}).reset_index()

# Rename columns for clarity
result_df.rename(columns={
    "normalised prices": "mean_normalised_prices",
    "converted_quantity": "total_converted_quantity"
}, inplace=True)

result_df
```

```python
from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt

# Initialize MinMaxScaler
scaler = MinMaxScaler()

# Apply scaling to both columns
result_df[["scaled_mean_normalised_prices", "scaled_total_converted_quantity"]] = scaler.fit_transform(
    result_df[["mean_normalised_prices", "total_converted_quantity"]]
)

# Convert 'DATE' column to datetime if not already in datetime format
result_df["DATE"] = pd.to_datetime(result_df["DATE"], errors="coerce")

# Filter the data for the two time periods
mask_1 = (result_df["DATE"] >= "2022-10-31") & (result_df["DATE"] <= "2023-10-31")
mask_2 = (result_df["DATE"] > "2023-10-31") & (result_df["DATE"] <= "2024-10-31")

result_df_2022_2023 = result_df[mask_1]
result_df_2023_2024 = result_df[mask_2]

# Plot for the first time period (31st Oct 2022 - 31st Oct 2023)
plt.figure(figsize=(15, 6))
plt.plot(result_df_2022_2023["DATE"], result_df_2022_2023["scaled_total_converted_quantity"], label="Scaled Total Converted Quantity")
plt.title("Time-Series of Scaled Metrics (31st Oct 2022 - 31st Oct 2023)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Scaled Values", fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot for the second time period (31st Oct 2023 - 31st Oct 2024)
plt.figure(figsize=(15, 6))
plt.plot(result_df_2023_2024["DATE"], result_df_2023_2024["scaled_total_converted_quantity"], label="Scaled Total Converted Quantity")
plt.title("Time-Series of Scaled Metrics (31st Oct 2023 - 31st Oct 2024)", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("Scaled Values", fontsize=12)
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot the box plot for 'INR_per_quintal' column to identify outliers
plt.figure(figsize=(10, 6))
sns.boxplot(data=result_df, x="INR_per_quintal")
plt.title("Boxplot of INR_per_quintal (Outliers Visualized)", fontsize=14)
plt.xlabel("INR_per_quintal", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Calculate the IQR (Interquartile Range) for 'INR_per_quintal'
Q1 = result_df["INR_per_quintal"].quantile(0.25)
Q3 = result_df["INR_per_quintal"].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from 'INR_per_quintal' column
merged_df_no_outliers = result_df[(result_df["INR_per_quintal"] >= lower_bound) & (result_df["INR_per_quintal"] <= upper_bound)]

# Fill missing values with the average of the previous and next values
merged_df_no_outliers["INR_per_quintal"] = merged_df_no_outliers["INR_per_quintal"].interpolate(method='linear')
merged_df_no_outliers
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(12, 6))
plt.plot(merged_df_no_outliers["DATE"], merged_df_no_outliers["INR_per_quintal"], label="INR_per_quintal", color='b')
plt.title("DATE vs INR_per_quintal", fontsize=14)
plt.xlabel("Date", fontsize=12)
plt.ylabel("INR per Quintal", fontsize=12)
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
