```python
import pandas as pd
df = pd.read_csv('Data/aggregated_spice_price_data_2019_2024.csv')
df
```

```python
df.columns
```

```python
import matplotlib.pyplot as plt
df['Price Date'] = pd.to_datetime(df['Price Date'])  # Automatically handles YYYY-MM-DD format
df.set_index('Price Date', inplace=True)

# Filter data for the date range from 01-01-2023 to 31-12-2023
start_date = '2023-01-01'
end_date = '2023-12-31'
filtered_df = df.loc[start_date:end_date]

# Plot the time series
plt.figure(figsize=(12, 6))
plt.plot(filtered_df.index, filtered_df['Modal Price (Rs./Quintal)'], label='Modal Price (Rs./Quintal)', color='blue')
plt.title('Modal Price Variation Throughout 2023', fontsize=16)
plt.xlabel('Date', fontsize=12)
plt.ylabel('Modal Price (Rs./Quintal)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
