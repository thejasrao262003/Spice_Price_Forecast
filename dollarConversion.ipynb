```python
import pandas as pd
df = pd.read_csv("Export.csv")
df
```

```python
dollar_price_df = pd.read_csv("USD_INR Historical Data.csv")
dollar_price_df
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%d-%m-%Y')
dollar_price_df['Date'] = pd.to_datetime(dollar_price_df['Date'], format='%d-%m-%Y')

# Merge df with dollar_price_df on the DATE and Date columns
merged_df = pd.merge(df, dollar_price_df[['Date', 'Price']],
                     left_on='DATE', right_on='Date', how='left')

# Forward fill missing Price values using .ffill() method
merged_df['Price'] = merged_df['Price'].ffill()

# Optionally, if you want the DATE column in the original string format
merged_df['DATE'] = merged_df['DATE'].dt.strftime('%d-%m-%Y')

# Create the final DataFrame with all columns from df and the Price column
final_df = merged_df[['DATE'] + [col for col in df.columns if col != 'DATE'] + ['Price']]

print(final_df)
```

```python
final_df.to_csv("ExportDollar.csv", index=False)
```
